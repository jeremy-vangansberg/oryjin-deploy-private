```python
import requests

# Format correct pour LangGraph - utilise directement le message comme input
response = requests.post(
    "https://ht-nautical-decoration-70-7d3b580d2cc25c4fb59221f2145e155e.us.langgraph.app/runs/stream",
    headers={
      "Content-Type": "application/json",
      "X-Api-Key": "***REMOVED***"
    },
    json={
      "assistant_id": "fe096781-5601-53d2-b2f6-0d3403f7e9ca",
      "input": {
          "messages": [
              {
                  "type": "human", 
                  "content": "Je souhaite créer une campagne marketing pour augmenter les ventes de mon e-commerce de mode. Ma cible principale est les femmes de 25-45 ans urbaines avec un pouvoir d'achat moyen à élevé."
              }
          ]
      }
    }
)
```

```python
# Exécutez cette cellule pour voir les détails de la réponse
print("Status Code:", response.status_code)
print("Raw Response:")
print(response.text)

# Parser les événements SSE pour une meilleure lisibilité
events = response.text.split('\n')
for event in events:
    if event.strip():
        print(f"-> {event}")
```

```python
# Test avec différents formats de requête
import requests

# Format 1: Simple string input
response1 = requests.post(
    "https://ht-nautical-decoration-70-7d3b580d2cc25c4fb59221f2145e155e.us.langgraph.app/runs/stream",
    headers={
      "Content-Type": "application/json",
      "X-Api-Key": "***REMOVED***"
    },
    json={
      "assistant_id": "fe096781-5601-53d2-b2f6-0d3403f7e9ca",
      "input": "Je souhaite créer une campagne marketing pour augmenter les ventes."
    }
)

print("=== FORMAT 1: Simple string ===")
print(f"Status: {response1.status_code}")
print(response1.text[:200] + "..." if len(response1.text) > 200 else response1.text)
print()
```
